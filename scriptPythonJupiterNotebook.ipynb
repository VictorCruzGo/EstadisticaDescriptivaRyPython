```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import NullFormatter 
```

```python
mu=0.5
sd=0.3
y=mu+sd*np.random.randn(1000)
y=y[(y>0) & (y<1)]
y.sort()
x=np.arange(len(y))
```

```python
plt.figure(figsize=(10,8)) #Cambia el tamaño del grafico 10x8 pixeles

plt.subplot(2,2,1)
plt.plot(x,y)
plt.yscale("linear")
plt.xscale("linear")
plt.title("Escala Lineal")
plt.grid(True) #Habilta las grillas de la figura

plt.subplot(2,2,2)
plt.plot(x,y)
plt.yscale("log")
plt.title("Escala Logaritmica")
plt.grid(True)

plt.subplot(2,2,3)
plt.plot(x,y-y.mean())
plt.yscale("symlog", linthreshy=0.01)
plt.title("Escala Log Simetrico")
plt.grid(True)

plt.subplot(2,2,4)
plt.plot(x,y)
plt.yscale("logit")
plt.title("Escala logistica")
plt.gca().yaxis.set_minor_formatter(NullFormatter()) #getcurrent axis, Obtiene el eje actual y quita formatos de la escala 'y'
plt.grid(True)

plt.subplots_adjust(top=0.92, bottom=0.08, left=0.1, right=0.95, hspace=0.35, wspace=0.35) #Aumentar o disminuir los tamaños de los graficos.

plt.show()
```

## Division de datos por columnas

```python
import pandas as pd
```

```python
df=pd.read_csv("C:/dataFrame/bodyfat.txt", delimiter=" ")
df
```

```python
L=[0,1,0,1,2,0] #Vector que definiar las variables de la agrupacion.
```

```python
df3=pd.DataFrame(
    {
        'key':['A','B','C','A','B','C'],
        'value1':[0,1,2,3,4,5],
        'value2':[4,41,46,48,26,52]
    }
)
```

```python
df3
```

```python
df3.groupby(L).sum() #El parametro del groupby no necesariamente es una variable del dataframe. En este caso el parametro del group by es un vector 'L'.
#Ej. Las posiciones donde se encuentra el valor '1'=(1+3)=4 y (41+48)=89 respectivamente.
```

```python
df3.groupby('key').sum()
```

```python
df3.groupby(df3['key']).sum()
```

```python
df2=df3.set_index('key')
df2
```

```python
mapping={
    'A':'vowel',
    'B':'consonant',
    'c':'consonnat'
}
```

```python
df2.groupby(mapping).sum() #similar al ejemplo 'L'.La agrupacion se realiza por la variable 'mapping'
```

```python
df2.groupby(str.lower).mean() #str.lower, es una funcion de python
```

```python
df2.groupby([str.lower, mapping]).mean() #Agrupacion por dos funciones. 
```

## Graficos desde dats frames y exportacion

```python
import numpy as np
years= [year for year in range(1990, 2020)]
deads=[(y + np.random.uniform(0,100)-1850) for y in years]
```

```python
df= pd.DataFrame({
    'year': years,
    'deads':deads
})
```

```python
df.head()
```

```python
df.plot(kind="scatter", x="year", y="deads", figsize=(10,8)) #A partir de un dataframe, generar un Grafico sin matplotlib
```

```python
my_plot=df.plot(x="year", y="deads", figsize=(10,8))
```

```python
my_fig=my_plot.get_figure() #Obtener la figura para exportar a formato png
```

```python
my_fig.savefig("Downloads/data_frame_plot.png")
```
